# 1. Introduction

## 1.1 Motivation, Concept, Origin

> "develop open-source software, open-standards, and services for interactive computing across dozens of programming languages"

- Based upon IPython (another great tool), spun off in 2014
- Most important supported languages: Julia, Python & R
- Based upon web technology, allowing for a platform-agnostic architecture
- Centered around sandboxed kernels (language interpreters)
- Web technology allows for a multi-user plattform (JupyterHub) 
- Integration of markdown allows for writing structured, publication ready texts.


## 1.2 Notebook basics

Juypter notebooks are divided into cells. These can be one of three types:

- code
- markdown
- raw

All work with Jupyter notebook is done within cells. The cells allow for structuring your document while simultaneously enabling the mixing code and structured text

When interacting with jupyter notebooks two modes of interaction can be distinguished:
- insert mode (green vertical bar on the left)
- global mode (bar turns blue).
